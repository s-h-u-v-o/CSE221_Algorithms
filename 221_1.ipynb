```python
n = int(input())
for i in range(n):
    s = input().split()
    if s[2] == '+':
        print(int(s[1])+int(s[3]))
    if s[2] == '-':
        print(int(s[1])-int(s[3]))
    if s[2] == '*':
        print(round(float(s[1])*float(s[3]), 6))
    if s[2] == '/':
        print(round(float(s[1])/float(s[3]), 6))
```

```python
def bubbleSort(arr):
    n = len(arr)
    flag = True
    while flag:
        flag = False
        for i in range(1, n):
            if int(arr[i-1]) > int(arr[i]):
                flag = True
                arr[i-1], arr[i] = arr[i], arr[i-1]
    return arr

n = int(input())
s = input().split()
arr = bubbleSort(s)
print(' '.join(arr))
```

```python
n = int(input())
arr = list(map(int, input().split()))
arr2 = list(map(int, input().split()))
for i in range(n):
    arr[i] = (arr[i], arr2[i])
count = 0
for i in range(n-1):
    max = i
    for j in range(i+1, n):
        if arr[max][1] < arr[j][1] or (arr[max][0] > arr[j][0] and arr[max][1] == arr[j][1]):
            max = j
    if max != i:
        arr[i], arr[max] = arr[max], arr[i]
        count += 1
print('Minimum swaps:', count)
for i in range(n):
    print(f'ID: {arr[i][0]} Mark: {arr[i][1]}')
```

```python
n = int(input())
lst = []
for i in range(n):
    lst.append(input().split()+[i])
for i in range(n):
    min = i
    for j in range(i+1, n):
        if lst[min][0] > lst[j][0]:
            min = j
        elif lst[min][0] == lst[j][0] and lst[min][6] < lst[j][6]:
            min = j
        elif lst[min][0] == lst[j][0] and lst[min][6] == lst[j][6] and lst[min][7] > lst[j][7]:
            min = j
    if i != min:
        lst[min], lst[i] = lst[i], lst[min]
for i in range(n):
    print(' '.join(lst[i][:7]))
```
