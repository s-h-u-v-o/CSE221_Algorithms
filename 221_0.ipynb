```python
def binary_search(arr, val):
    left, right = 0, len(arr)-1
    while left <= right:
        mid = left + (right-left)//2
        if arr[mid] == val:
            return True
        else:
            if arr[mid] < val:
                left = mid+1
            else:
                right = mid-1
    return False


arr = [2, 5, 6, 9, 12, 16, 34, 45, 56, 67, 78, 89, 90]
binary_search(arr, 14)


# to find the index of the element (where it is or where it should be):

# def binary_search(arr, val, left=0, right=len(arr)):
#     left, right = left, right
#     while left < right:
#         mid = left + (right-left)//2
#         if arr[mid] < val:
#             left = mid+1
#         else:
#             right = mid
#     return left
```

```python
def bubble_sort(arr):
    n = len(arr)
    flag = True
    while flag:
        flag = False
        for i in range(1, n):
            if arr[i-1] > arr[i]:
                flag = True
                arr[i-1], arr[i] = arr[i], arr[i-1]


arr = [4, 23, 45, 445, 65, 435, 67, 668, 32, 14, 2, 323, 134, 23, 567, 76]
print(arr)
bubble_sort(arr)
print(arr)
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key


arr = [4, 23, 45, 445, 65, 435, 67, 668, 32, 14, 2, 323, 134, 23, 567, 76]
print(arr)
insertion_sort(arr)
print(arr)
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n-1):
        min = i
        for j in range(i+1, n):
            if arr[j] < arr[min]:
                min = j
            if min != i:
                arr[i], arr[min] = arr[min], arr[i]


arr = [4, 23, 45, 445, 65, 435, 67, 668, 32, 14, 2, 323, 134, 23, 567, 76]
print(arr)
selection_sort(arr)
print(arr)
```

```python
def partition(array, low, high):

    # choose the rightmost element as pivot
    pivot = array[high]
    # pointer for greater element
    i = low - 1
    for j in range(low, high):
        if array[j] <= pivot:
            i = i + 1
            (array[i], array[j]) = (array[j], array[i])
    (array[i + 1], array[high]) = (array[high], array[i + 1])
    return i + 1


def quickSort(array, low, high):
    if low < high:
        pi = partition(array, low, high)
        quickSort(array, low, pi - 1)
        quickSort(array, pi + 1, high)


data = [1, 2, 4, 5, 7, 3, 8, 0, 9]
print("Unsorted Array")
print(data)

quickSort(data, 0, len(data) - 1)

print('Sorted Array in Ascending Order:')
print(data)
```

```python
def maxSubarraySum(arr, l, r):
    if l > r:
        return float('-inf')
    if l == r:
        # Base case: one element
        return arr[l]
    # Find the middle point
    m = l + (r-l)//2
    # Return the maximum of three cases
    return max(maxSubarraySum(arr, l, m), maxSubarraySum(arr, m + 1, r), maxCrossingSum(arr, l, m, r))


def maxCrossingSum(arr, l, m, r):
    # Include elements on the left of mid
    leftSum = float('-inf')
    sum = 0
    for i in range(m, l - 1, -1):
        sum += arr[i]
        leftSum = max(leftSum, sum)
    # Include elements on the right of mid
    rightSum = float('-inf')
    sum = 0
    for i in range(m, r + 1):
        sum += arr[i]
        rightSum = max(rightSum, sum)
    # Return the maximum of left sum, right sum, and their combination
    return max(leftSum + rightSum - arr[m], leftSum, rightSum)


arr = [2, 3, 4, 5, 7]
print(maxSubarraySum(arr, 0, len(arr)-1))
```

```python
def Kadane(arr):
    result = maxEnding = arr[0]
    for i in range(1, len(arr)):
        maxEnding = max(maxEnding + arr[i], arr[i])
        result = max(result, maxEnding)
    return result

arr = [2, 3, -8, 7, -1, 2, 3]
print(Kadane(arr))
```

```python
def karatsuba(x, y):
    if x<10 or y<10:
        return x*y
    n = max(len(str(x)), len(str(y)))
    m = n//2
    p = 10**m
    xh = x//p
    xl = x%p
    yh = y//p
    yl = y%p
    a = karatsuba(xh, yh)
    b = karatsuba(xl, yl)
    c = karatsuba(xh+xl, yh+yl)
    return a*10**(2*m) + (c-a-b)*10**m + b

print(karatsuba(123, 23456))
print(123*23456)
```
