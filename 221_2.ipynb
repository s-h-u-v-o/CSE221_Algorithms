```python
def two_sum():
    n, sum = map(int,input().split())
    arr = list(map(int,input().split()))
    i= 1
    j = n
    while i != j:
        if arr[i-1] + arr[j-1] == sum:
            return str(i) + ' ' + str(j)
        elif arr[i-1] + arr[j-1] < sum:
            i += 1
        else:
            j-=1
    return -1
print(two_sum())
```

```python
m = int(input())
l1 = [i for i in input().split()]
n = int(input())
l2 = [i for i in input().split()]
l3 = [0]*(m+n)
while m>0 and n>0:
    if int(l1[m-1]) >= int(l2[n-1]):
        l3[m+n-1] = l1[m-1]
        m -= 1
    else:
        l3[m+n-1] = l2[n-1]
        n -= 1
while m == 0 and n != 0:
    l3[m+n-1] = l2[n-1]
    n -= 1
while n == 0 and m != 0:
    l3[m+n-1] = l1[m-1]
    m -= 1
print(' '.join(l3))
```

```python
n, k = map(int, input().split())
arr = [int(i) for i in input().split()]
i, mx = 0, 0
sum = 0
for j in range(n):
    sum += arr[j]
    while sum > k and i <= j:
        sum -= arr[i]
        i += 1
    mx = max(mx, j-i+1)
print(mx)
```

```python
n, k = map(int, input().split())
arr = [int(i) for i in input().split()]
for i in range(k):
    x, y = -1, -1
    a, b = map(int, input().split())
    left, right = 0, n-1
    while left <= right:
        mid = left + (right-left)//2
        if arr[mid] == a:
            x = mid
            right = mid-1
        else:
            if arr[mid] < a:
                left = mid+1
            else:
                right = mid-1
    if x == -1:
        x = left
    left, right = 0, n-1
    while left <= right:
        mid = left + (right-left)//2
        if arr[mid] == b:
            y = mid
            left = mid+1
        else:
            if arr[mid] < b:
                left = mid+1
            else:
                right = mid-1
    if y == -1:
        y = right
    print(y-x+1)

# bisect module
```
