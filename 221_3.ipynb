```python
def merge_sort(arr, count=0):
    n = len(arr)
    if n <= 1:
        return arr,0
    m = n//2
    left, c1 = merge_sort(arr[:m])
    right, c2 = merge_sort(arr[m:])
    arr1, c = merge(left, right)
    return arr1, c+c1+c2
def merge(left, right):
    arr2 = []
    count = 0
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            arr2.append(left[i])
            i += 1
            count += 1
        else:
            arr2.append(right[j])
            j += 1
    arr2.extend(left[i:])
    arr2.extend(right[j:])
    return arr2, count

n = int(input())
arr = [int(i) for i in input().split()]
arr1, count = merge_sort(arr)
print(count)
print(' '.join(map(str, arr1)))
```

```python
def pair_max(arr):
    n = len(arr)
    if n < 2:
        return None
    curr_mx = float('-inf')
    mx = arr[0]
    for i in range(1, n):
        curr_mx = max(curr_mx, mx + arr[i]**2)
        mx = max(mx, arr[i])
    return curr_mx

x = int(input())
arr = [int(i) for i in input().split()]
print(pair_max(arr))
```

```python
def fast_mod(a, b):
    rslt = 1
    mod = 107
    a = a % mod
    while b > 0:
        if b%2 == 1:
            rslt = (rslt*a) % mod
        a = (a*a) % mod
        b = b//2
    return rslt

a, b = map(int, input().split())
print(fast_mod(a, b))
```

```python
def mod_sum_of_powers(a, n, m):
    # Step 1: Detect the cycle
    seen = {}
    cycle = []
    current = 1  # Initial power of a^1
    while current not in seen:
        seen[current] = len(cycle)
        cycle.append(current)
        current = (current * a) % m

    # Step 2: Break the sequence into pre-cycle and cycle components
    cycle_start = seen[current]
    cycle_sum = sum(cycle[cycle_start:])
    pre_cycle_sum = sum(cycle[:cycle_start])
    cycle_length = len(cycle) - cycle_start

    # Step 3: Calculate the total sum
    if n <= len(cycle):  # If n is small enough, no need for full cycles
        return sum(cycle[:n]) % m

    total_sum = pre_cycle_sum
    n -= cycle_start  # Exclude the pre-cycle part
    total_sum += (n // cycle_length) * cycle_sum  # Add full cycles
    total_sum += sum(cycle[cycle_start:cycle_start + (n % cycle_length)])  # Add remaining terms

    return total_sum % m

# Input and execution
T = int(input())
results = []
for _ in range(T):
    a, n, m = map(int, input().split())
    results.append(mod_sum_of_powers(a, n, m))

# Output the results
print("\n".join(map(str, results)))
```

```python
def minimize_bst_height(arr, s, e):
    if s > e:
        return []
    m = s + (e-s)//2
    left = minimize_bst_height(arr, s, m-1)
    right = minimize_bst_height(arr, m+1, e)
    return [arr[m]] + left + right

n = int(input())
arr = [int(i) for i in input().split()]
arr = minimize_bst_height(arr, 0, n-1)
print(' '.join(map(str, arr)))
```

```python
def find_post_order(in_order, pre_order):
    # If the in-order or pre-order is empty, there's no tree to process.
    if not in_order or not pre_order:
        return []

    # The first element of pre-order is always the root of the tree.
    root = pre_order[0]

    # Find the root in the in-order traversal to split into left and right subtrees.
    root_index = in_order.index(root)

    # Recursively compute post-order for the left and right subtrees.
    left_post_order = find_post_order(in_order[:root_index], pre_order[1:1 + root_index])
    right_post_order = find_post_order(in_order[root_index + 1:], pre_order[1 + root_index:])

    # Post-order is: left subtree, right subtree, root.
    return left_post_order + right_post_order + [root]

# Input Handling
n = int(input())
in_order = list(map(int, input().split()))
pre_order = list(map(int, input().split()))

# Find and print the post-order traversal
post_order = find_post_order(in_order, pre_order)
print(" ".join(map(str, post_order)))
```
