```python
n, m = map(int, input().split())
g = [[0] * n for i in range(n)]
for i in range(m):
    a, b, c = map(int, input().split())
    g[a - 1][b - 1] = c
for i in range(n):
    print(*g[i])
```

```python
n, m = map(int, input().split())
a = [int(i) for i in input().split()]
b = [int(i) for i in input().split()]
c = [int(i) for i in input().split()]
d = {}
for i in range(m):
    if a[i] not in d:
        d[a[i]] = []
    d[a[i]].append((b[i], c[i]))
for i in range(1, n + 1):
    print(f'{i}:', *d[i]) if i in d else print(f'{i}:')
```

```python
n = int(input())
g = [[0] * n for _ in range(n)]
for i in range(n):
    a = [int(x) for x in input().split()]
    for j in range(a[0]):
        g[i][a[j+1]] = 1
for i in range(n):
    print(*g[i])
```

```python
n, m = map(int, input().split())
a = [int(x) for x in input().split()]
b = [int(x) for x in input().split()]
deg = [0]*n
for i in range(m):
    deg[a[i]-1] += 1
    deg[b[i]-1] += 1
odd = 0
for i in range(n):
    if deg[i] % 2 == 1:
        odd += 1
if odd == 0 or odd == 2:
    print("YES")
else:
    print("NO")
```

```python
n, m = map(int, input().split())
a = [int(x) for x in input().split()]
b = [int(x) for x in input().split()]
indeg, outdeg = [0] * n, [0] * n
for i in range(m):
    indeg[a[i] - 1] += 1
    outdeg[b[i] - 1] += 1
for i in range(n):
    indeg[i] = outdeg[i] - indeg[i]
print(*indeg)
```

```python
def gcd(x, y):
    while y:
        x, y = y, x % y
    return x
n, m = map(int, input().split())
a = []
for i in range(1, n+1):
    b = []
    for j in range(1, n+1):
        if gcd(i, j) == 1 and i != j:
            b.append(j)
    a.append(b)
for i in range(m):
    x, y = map(int, input().split())
    print(a[x-1][y-1] if y <= len(a[x-1]) else -1)
```
